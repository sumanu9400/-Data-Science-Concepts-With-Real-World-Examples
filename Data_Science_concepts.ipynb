```python
import numpy as np

# Define matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# 1. Matrix Multiplication
print("\n1. Matrix Multiplication (A @ B):\n", A @ B)

# 2. Dot Product
print("\n2. Dot Product (A dot B):\n", np.dot(A, B))

# 3. Transpose
print("\n3. Transpose of A:\n", A.T)

# 4. Determinant
print("\n4. Determinant of A:", np.linalg.det(A))

# 5. Inverse
print("\n5. Inverse of A:\n", np.linalg.inv(A))

# 6. Eigenvalues & Eigenvectors
eigen_vals, eigen_vecs = np.linalg.eig(A)
print("\n6. Eigenvalues:", eigen_vals)
print("   Eigenvectors:\n", eigen_vecs)

# 7. Norm of a Matrix
print("\n7. Matrix Norm (||A||):", np.linalg.norm(A))

# 8. Solve Linear Equation Ax = b
b = np.array([5, 6])
x = np.linalg.solve(A, b)
print("\n8. Solution of Ax = b:", x)
```

✅ 1. Hierarchical Indexing (MultiIndex)
Real-Life Example: Sales data for different cities & months


```python
import pandas as pd

sales = pd.Series(
    [1200, 1500, 1800, 2200],
    index=[['Hyderabad','Hyderabad','Chennai','Chennai'],
           ['Jan','Feb','Jan','Feb']]
)

print(sales)
```

Selecting Chennai-Feb sales

```python
sales.loc[('Chennai','Feb')]
```

✅ 2. merge() — SQL Style Join
Real-Life Example: Customer table + Orders table

```python
customers = pd.DataFrame({
    'cust_id':[101,102,103],
    'name':['Ravi','Sita','Kiran']
})

orders = pd.DataFrame({
    'cust_id':[101,103,103],
    'order_amount':[500,1200,800]
})

merged = pd.merge(customers, orders, on='cust_id', how='inner')
print(merged)
```

✅ 3. join() — Join Based on Index
Real-Life Example: Student marks stored with index


```python
df1 = pd.DataFrame({'math':[90,85,78]}, index=['A','B','C'])
df2 = pd.DataFrame({'science':[88,92,80]}, index=['A','B','C'])

res = df1.join(df2)
print(res)
```

✅ 4. concat() — Combine Datasets (Add rows or columns)


A. Row-wise concatenation
Real-life: Adding new monthly sales.

```python
jan = pd.DataFrame({'sales':[1000, 1200]}, index=['Store1','Store2'])
feb = pd.DataFrame({'sales':[1300, 1250]}, index=['Store1','Store2'])

res = pd.concat([jan, feb])
print(res)
```

B. Column-wise concatenation

```python
temp = pd.DataFrame({'temp':[35, 32]}, index=['Hyd','Chennai'])
humidity = pd.DataFrame({'hum':[60, 58]}, index=['Hyd','Chennai'])

res = pd.concat([temp, humidity], axis=1)
print(res)
```

✅ 5. pivot() — Reshaping Data (wide format)


Real-Life Example: Temperature recorded for cities

```python
import pandas as pd

df = pd.DataFrame({
    'city':['Hyd','Hyd','Chennai','Chennai'],
    'month':['Jan','Feb','Jan','Feb'],
    'temp':[30, 32, 28, 29]
})

pivoted = df.pivot(index='month', columns='city', values='temp')
print(pivoted)
```

✅ 6. pivot_table() — Pivot + Aggregation


Real-Life Example: Multiple sales entries per month

```python
df = pd.DataFrame({
    'city':['Hyd','Hyd','Hyd','Chennai'],
    'month':['Jan','Jan','Feb','Jan'],
    'sales':[200,300,250,400]
})

pt = df.pivot_table(index='city', columns='month', values='sales', aggfunc='sum')
print(pt)
```

✅ 7. melt() — Wide → Long Format


Real-Life Example: Student marks

```python
df = pd.DataFrame({
    'name':['Ravi','Sita'],
    'math':[80, 90],
    'science':[85, 95]
})

melted = pd.melt(df, id_vars=['name'], var_name='subject', value_name='marks')
print(melted)
```

✅ 8. stack() / unstack()

Real-Life Example: Sales with MultiIndex

```python
df = pd.DataFrame(
    [[1200, 1500], [1800, 1600]],
    index=['Hyd','Chennai'],
    columns=['Jan','Feb']
)
```

```python
print(df.unstack())
```

✅ 5. Data Aggregation and Group Operations

Grouping allows you to perform operations on groups of data.

⭐ GroupBy Mechanics
Example Dataset

```python
import pandas as pd

df = pd.DataFrame({
    "team": ["A", "A", "B", "B", "B"],
    "sales": [100, 150, 200, 50, 70]
})
```

Grouping

```python
grouped = df.groupby("team")
```

⭐ Data Aggregation

Common aggregations:

```python
df.groupby("team")["sales"].sum()
```

```python
df.groupby("team")["sales"].agg(["mean", "max", "min"])
```

⭐ Apply: Split–Apply–Combine

Used when aggregation isn’t enough.

Example:

Apply 10% discount to each team’s average sales:

```python
def discount(x):
    return x.mean() * 0.9

df.groupby("team")["sales"].apply(discount)
```

⭐ Pivot Tables

Pivot tables summarize data in table form.

Example:

Sales by team and category

```python
df = pd.DataFrame({
    'team': ['A','A','B','B'],
    'category': ['Food','Tech','Food','Tech'],
    'sales': [100,200,150,220]
})

pd.pivot_table(df, values='sales', index='team', columns='category', aggfunc='sum')
```

⭐ Cross-Tabulation

Used for counting categories.

Example:

```python
pd.crosstab(df['team'], df['category'])
```
